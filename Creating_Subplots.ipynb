```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
ax1 = plt.axes() #Create the main axes
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2]) #Create an inner subplot / inset axes
```

```python
#Create an inner subplot using the object-oriented interface.
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                  xticklabels = [],
                  ylim = (-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                  ylim = (-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.cos(x))
ax2.plot(np.sin(x));
```

```python
#Create a simple Grid of Subplots
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
            fontsize = 18, ha = 'center')
```

```python
#Create a simple Grid of Subplots with the equivalent object-oriented command.
fig = plt.figure()
fig.subplots_adjust(hspace = 0.4, wspace = 0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
            fontsize = 18, ha = 'center')
```

```python
#Create the Grid in one line.
fig, ax = plt.subplots(2, 3, sharex = 'col', sharey = 'row')

#Add lables.
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                     fontsize = 18, ha = 'center')
```

```python
grid = plt.GridSpec(2, 3, wspace = 0.4, hspace = 0.3)
```

```python
#Create mutliple subplots of various sizes.
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

```python
#Visualize multidimensional Distributions
```

```python
mean = [0, 0]
```

```python
cov = [[1, 1], [1, 2]]
```

```python
x, y = np.random.multivariate_normal(mean, cov, 3000).T
```

```python
#Set up the axes
```

```python
fig = plt.figure(figsize = (6, 6))
grid = plt.GridSpec(4, 4, hspace = 0.2, wspace = 0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels = [], sharey = main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels = [], sharex = main_ax)

#Display a scatter plot on the main axes.
main_ax.plot(x, y, 'ok', markersize = 3, alpha = 0.2)

#Histogram on the attached axes
x_hist.hist(x, 40, histtype = 'stepfilled',
           orientation = 'vertical', color = 'gray')
x_hist.invert_yaxis()
y_hist.hist(y, 40, histtype = 'stepfilled',
           orientation = 'horizontal', color = 'gray')
y_hist.invert_xaxis()
```
