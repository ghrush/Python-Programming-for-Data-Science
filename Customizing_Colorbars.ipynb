```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
```

```python
#Create an array of values.
x = np.linspace(0, 10, 1000)
```

```python
#Create another array for ploting data based on the orginal array.
I = np.sin(x) * np.cos(x[:, np.newaxis])
```

```python
#Display a simple colorbar.
plt.imshow(I)
plt.colorbar();
```

```python
#Display the colormap in grayscale.
plt.imshow(I, cmap = 'gray')
```

```python
##Examining Colormaps and its Grayscale Equivalents
```

```python
#This function returns a grayscale of a given map.
def grayscale_cmap(cmap):
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    RGB_weight = [0.299, 0.587, 0.114]
    luminance = np.sqrt(np.dot(colors[:, :3] **2, RGB_weight))
    colors[:, :3] = luminance[:, np.newaxis]
    
    return LinearSegmentedColormap.from_list(cmap.name + "_gray", colors, cmap.N)
    
```

```python
#This function returns a colormap with its grayscale equivalent.
def view_colormap(cmap):
    
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    cmap = grayscale_cmap(cmap)
    grayscale = cmap(np.arange(cmap.N))
    
    fig, ax = plt.subplots(2, figsize = (6, 2),
                          subplot_kw = dict(xticks = [], yticks = []))
    ax[0].imshow([colors], extent = [0, 10, 0, 1])
    ax[1].imshow([grayscale], extent = [0, 10, 0, 1])
```

```python
#Display the 'jet' colormap and its grayscale equivalent.
view_colormap('jet')
```

```python
#Display the 'viridis' colormap and its grayscale equivalent.
view_colormap('viridis')
```

```python
#Display the 'cubehelix' colormap and its grayscale equivalent.
view_colormap('cubehelix')
```

```python
#Display the 'RdBu' colormap and its grayscale equivalent.
view_colormap('RdBu')
```
