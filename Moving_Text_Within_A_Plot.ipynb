```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(facecolor = 'lightgray')
ax.axis([0, 10, 0, 10])

#Place text on plot using transData.
ax.text(1, 5, ". Data: (1, 5)", transform = ax.transData)

#Place text on plot using transAxes.
ax.text(0.5, 0.1, ". Data: (0.5, 0.1)", transform = ax.transAxes)
```

```python
#Modify the axes limts.
ax.set_xlim(0, 2)
ax.set_ylim(-6, 6)

#Display the plot.
fig
```
