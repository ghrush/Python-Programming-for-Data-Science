```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
```

```python
#Generate an array of 1000 random values.
x = np.linspace(0, 10, 1000)
```

```python
#Plot the sine and cosine of the array of values with a legend.
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label = 'Sine')
ax.plot(x, np.cos(x), '--r', label = 'Cosine')
ax.axis('equal')
leg = ax.legend();
```

```python
#Move the legend to the upper left corner of the plot.
ax.legend(loc = 'upper left', frameon = False)
fig
```

```python
#Specifiy the columns in the legend.
ax.legend(loc = 'lower center', frameon = False, ncol = 2)
fig
```

```python
#Place a frame around the legend.
ax.legend(fancybox = True, framealpha = 1, shadow = True, borderpad = 1)
fig
```

```python
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
```

```python
#Plot multiple lines at once. Select with lines to display in the legend.
lines = plt.plot(x, y)
plt.legend(lines[:3], ['first', 'second', 'third'])
```

```python
#Use alternate method of displaying legend elements.
plt.plot(x, y[:, 0], label = 'first')
plt.plot(x, y[:, 1], label = 'second')
plt.plot(x, y[:, 2], label = 'third')
plt.plot(x, y[:, 3], label = 'fourth')
plt.legend(framealpha = 1, frameon = True);
```
